```python
############read from jason file
import pandas as pd
import numpy as np
import json,xlwt,sys
from datetime import datetime
from pandas.io.json import json_normalize


with open('week3_homework.json') as f:
    txt = f.readlines()

for line in txt:
    s=line.split('eventSequence') [1] 
    
s1=s[3:-4]
s1=s1.replace('\"', '\'')
s2=s1.split('},{')
s4={}
for i in s2:
    i=i[1:]
    for j in i.strip().split(','):
        k=j.strip().split(':')
        s4.setdefault(k[0].replace('\'',''),[]).append(k[1].replace('\'',''))
print("event serial is: ", s4)

columnname=list(s4.keys())
df = pd.DataFrame(s4, columns=columnname)
df.to_excel('tmp1.xlsx', index=False, header=True)
```
